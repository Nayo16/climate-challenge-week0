```python
import pandas as pd

ethiopia = pd.read_csv('../data/ethiopia_clean.csv')
kenya = pd.read_csv('../data/kenya_clean.csv')
sudan = pd.read_csv('../data/sudan_clean.csv')
tanzania = pd.read_csv('../data/tanzania_clean.csv')
nigeria = pd.read_csv('../data/nigeria_clean.csv')

df = pd.concat([ethiopia, kenya, sudan, tanzania, nigeria])
```

## Cross-Country Climate Comparison
This notebook combines cleaned climate data from Ethiopia, Kenya, Sudan, Tanzania, and Nigeria to provide a comparative regional analysis. The objective is to identify extreme heat events, compare monthly temperature patterns, and assess rainfall distribution across the five countries.

```python
df[df["T2M_MAX"] > 35]
```

```python
# Temperature comparison - line chart for all countries
import matplotlib.pyplot as plt
monthly_temp_by_country = df.groupby(['Country', 'Month'])['T2M'].mean().unstack(level=0)
monthly_temp_by_country.plot(figsize=(12, 6))
plt.title('Monthly Average Temperature Comparison Across Countries')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.legend(title='Country')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Rainfall comparison - boxplot
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Country', y='PRECTOTCORR')
plt.title('Rainfall Distribution Comparison Across Countries')
plt.xlabel('Country')
plt.ylabel('Precipitation (mm)')
plt.xticks(rotation=45)
plt.show()
```
